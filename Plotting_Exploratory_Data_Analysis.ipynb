### To set up a whitegrid for status domain of data

```python


sns.set_style("whitegrid");
sns.FacetGrid(data, hue="status", size=6) \
   .map(plt.scatter, "nodes", "age") \
   .add_legend();
plt.title('white grid status of data')
plt.show();
```

# 6.Perform Univaraite analysis(PDF, CDF, Boxplot, Voilin plots) to understand which features are useful towards classification.

### Histogram, PDF, CDF

### To apply a FaceGrid on status with displot of age

```python

sns.FacetGrid(data, hue="status", size=6) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.title('Face grid on status with displot of age')
plt.show();
```

### To apply a FaceGrid on status with displot of year

```python

sns.FacetGrid(data, hue="status", size=6) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.title('FaceGrid on status with displot of year')
plt.show();
```

### To apply a FaceGrid on status with displot of nodes

```python

sns.FacetGrid(data, hue="status", size=6) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.title('FaceGrid on status with displot of nodes')
plt.show();
```

### Commulative Distributive function of status of nodes 'y'

```python


counts1, bin_edges1 = np.histogram(status_positive['nodes'], bins=10, density = True)
pdf1 = counts1/(sum(counts1))
print(pdf1);
print(bin_edges1)
cdf1 = np.cumsum(pdf1)
plt.plot(bin_edges1[1:], pdf1)
plt.plot(bin_edges1[1:], cdf1, label = 'Yes')
plt.xlabel('nodes')
plt.legend()
plt.title('Commulative distributive function of status of nodes y')
plt.show()
```

### Observation

83.55% of the patients who have survived had nodes in the range of 0–4.6

### Commulative Distributive function of status of nodes 'n'

```python


counts2, bin_edges2 = np.histogram(status_negetive['nodes'], bins=10, density = True)
pdf2 = counts2/(sum(counts2))
print(pdf2);
print(bin_edges2)
cdf2 = np.cumsum(pdf2)
plt.plot(bin_edges2[1:], pdf2)
plt.plot(bin_edges2[1:], cdf2, label = 'No')
plt.xlabel('nodes')
plt.legend()
plt.title('Commulative distributive function of status of nodes n')
plt.show()
```

### Observation

56.79% of the patients who have survived had nodes in the range of 0–5.2
